```python
#import modules
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline  
plt.style.use('ggplot')

#read and change data type
df=pd.read_csv('covid_19_data.csv')
df=df.drop(columns=['Last Update'])
df['ObservationDate']=pd.to_datetime(df['ObservationDate'])

#filter on China, US and Italy
filt_countries=df['Country/Region'].isin(['Mainland China','US','Italy'])
ncov_countries=df[filt_countries].groupby(['ObservationDate','Country/Region'])[['Confirmed', 'Deaths', 'Recovered']].sum().reset_index()

#plot data
columns=['Confirmed', 'Deaths', 'Recovered']
plt.plot(ncov_countries['ObservationDate'],ncov_countries[columns])
plt.legend(['Confirmed', 'Deaths', 'Recovered'])
plt.title('CoronaVirus_China')
plt.xlabel('Date')
plt.ylabel('Number of People')
plt.tight_layout()



```
